## 학습에 사용할 열 선택
 - Y1 = load
 - Y2 = heating_load
 - Y3 = cooling_load
 - X = load, heating_load, cooling_load, relative_compactness, roof_area, orientation을 제외한 나머지 열

```python
df.drop(columns=["load", "heating_load", "cooling_load", "relative_compactness", "roof_area", "orientation"])
```

```python
# 코드를 입력해 주세요
# X와 Y를 분리하세요.
X = df.drop(columns=["load", "heating_load", "cooling_load", "relative_compactness", "roof_area", "orientation"])
Y1 = df["load"]
Y2 = df["heating_load"]
Y3 = df["cooling_load"]
```

 - X의 상위 5개 행과 shape을 출력하세요.

```python
# 코드를 입력해 주세요
print(df.shape)
X.head()
```

## 데이터 분할
 - 데이터를 학습 및 평가 데이터로 분할하세요.
 - 분할 비율은 75:25입니다.
 - random_state=0으로 설정하세요.

```python
# 코드를 입력해 주세요
from sklearn.model_selection import train_test_split


x_train, x_test,y1_train, y1_test, y2_train, y2_test, y3_train, y3_test =  train_test_split(X,Y1,Y2,Y3, test_size=0.25, random_state=0)

```

 - 분할된 데이터의 shape을 출력하세요.

```python
print(x_train.shape)
print(x_test.shape)
print(y1_train.shape)
print(y1_test.shape)
print(y2_train.shape)
print(y2_test.shape)
print(y3_train.shape)
print(y3_test.shape)
```

## 모델 생성 및 학습
 - LinearRegression 모델을 이용하여 모델링 후, 학습 및 평가 데이터의 R-squared 지표를 출력하세요.

```python
# 코드를 입력해 주세요
# load 학습
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(x_train,y1_train)
model1.score(x_train, y1_train), model1.score(x_test, y1_test)
```

```python
# 코드를 입력해 주세요
# heating load 학습
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
model2.fit(x_train, y2_train)
model2.score(x_train, y2_train), model2.score(x_test, y2_test)
```

```python
# 코드를 입력해 주세요
# cooling load 학습
from sklearn.linear_model import LinearRegression

model3 = LinearRegression()
model3.fit(x_train, y3_train)
model3.score(x_train, y3_train), model3.score(x_test, y3_test)
```

 - 각 학습된 모델의 회귀 계수와 절편을 출력하세요.

```python
print(model1.coef_, model1.intercept_)
print(model2.coef_, model2.intercept_)
print(model3.coef_, model3.intercept_)
```

## 모델 예측 및 평가
 - 전체 X 데이터의 예측 결과를 각각 heating_pred, cooling_pred, load_pred 열에 추가하세요.

```python
# 코드를 입력해 주세요
df['heating_pred'] = model2.predict(X)
df['cooling_pred'] = model3.predict(X)
df['load_pred'] = model1.predict(X)
df
```

 - 실제 load와 예측 load의 MSE, RMSE, RMSLE, MAE, MAPE를 각각 출력하세요.

```python
# 코드를 입력해 주세요
from sklearn.metrics import mean_squared_error, mean_squared_log_error, mean_absolute_error, mean_absolute_percentage_error

print(mean_squared_error(df["load"], df["load_pred"]))
print(mean_squared_log_error(df["load"], df["load_pred"]))
print(mean_absolute_error(df["load"], df["load_pred"]))
print(mean_absolute_percentage_error(df["load"], df["load_pred"]))
```

 - 실제 load가 예측 load보다 크면 'insufficient', 작거나 같으면 'sufficient'라는 값을 갖는 열 'state'를 추가하세요.
 - sufficient와 insufficient의 개수를 출력하세요.

```python
df["state"] = 0
df
```

```python
# 코드를 입력해 주세요
df["state"] =np.where(df["load"] > df["load_pred"], "insufficient" , "sufficient")
df
```
