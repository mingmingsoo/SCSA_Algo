## 연산

```python
x = np.array([[1,2], [3,4]])
y = np.array([[5,6], [7,8]])
x, y
```

```python
# 두 배열의 덧셈
x+y
np.add(x,y)
```

```python
# 두 배열의 뺄셈
x-y
np.subtract(x,y)
```

```python
# 두 배열의 곱셈
x*y
np.multiply(x,y)
```

```python
# 두 배열의 나눗셈
x/y
np.divide(x,y)
```

```python
# scalar 연산도 가능
x/3
```

```python
# scalar 연산도 가능
y*2
```

## Indexing

정수 인덱싱 : array[0], array[3, 3]  

조건 인덱싱 : array[조건] # 조건은 True or False 값을 가짐  

팬시 인덱싱 : array[row, col] - array[[0,1], [2,1]] 처럼 다른 array이용

### 정수 인덱싱

```python
# 0 ~ 8의 값을 갖는 2차원 배열 생성 - arange, reshape 사용

array = np.arange(9).reshape(3,3)
array
```

```python
# arr의 첫 번째 행 indexing하여 출력
array[0]
```

```python
# arr의 마지막 행(음수 index 사용) indexing하여 출력

array[-1]
```

```python
# arr의 첫 번째 행, 첫 번째 열 indexing 하여 출력
array[0][0]
```

### 조건 인덱싱

```python
# 1 ~ 5의 값을 갖는 1차원 배열 생성 - arange 사용

array = np.arange(1,6)
array
```

```python
# arr 배열에서 홀수의 값만 선택
for x in array:
    if(x%2==1):
        print(x)
bidx = array % 2 ==1
bidx
```

```python
array[bidx]
```

```python
# arr 배열에서 값이 3이상인 것만 선택
array[array>=3]
```

### 팬시 인덱싱

다른 array를 이용하는 인덱싱.  
조건 인덱싱도 True, False값을 갖는 array를 이용하므로 팬시 인덱싱의 일종이라고 할 수 있음.

```python
# 0 ~ 8의 값을 갖는 2차원 배열 생성

array = np.arange(0,9).reshape((3,3))
array
```

```python
# 0행 2열, 1행 1열의 값을 추출 - row = [], col = []

row = [0,1]
col = [2,1]
array[row,col]
```

## Slicing

`:` 을 사용하여 특정 범위 값에 접근

슬라이싱하여 새로운 배열을 만들 경우, 새로운 배열을 수정하면 원본 배열이 수정됨

방지)  
copy() 함수 혹은 팬시인덱싱 사용

```python
# 0~8을 갖는 (3,3) shape의 array 생성
array = np.arange(0,9).reshape((3,3))
array
```

```python
# a 변수에 1 행, 전체 열 추출
a = array[1]
a
```

```python
# a[1]을 1000으로 수정
a[1] = 1000
```

```python
# array값 확인
array
```

### copy 함수 사용

```python
# b 변수에 0 행, 전체 열 추출
b = array[0].copy()
b
```

```python
# b[1]을 2000으로 수정
b[1] = 2000
```

```python
# array값 확인
array
```

## 다양한 numpy 함수

### np.where

```python
# 0~8을 갖는 (3,3) shape의 array 생성
array = np.arange(0,9).reshape((3,3))
array
```

```python
# array에서 짝수인 요소의 위치를 반환

np.where(array%2==0)
```

```python
# array에서 짝수이면 True, 홀수면 False로 하는 array 생성

np.where(array%2==0, True,False)
```

### 통계 함수

`axis=0` 행과 행의 연산  
`axis=1` 열과 열의 연산  

min, max, mean, median, quantile, std, var, corrcoef, argmin, argmax, 
cumsum, cumprod

```python
# 0~8을 갖는 (3,3) shape의 array 생성
array = np.arange(9).reshape((3,3))
array
```

```python
# array 행과 행의 합
array.sum()
array.sum(axis=0)
```

```python
# array 열과 열의 합
array.sum(axis=1)
```

```python
# array 행과 행의 평균
array.mean(axis=0)
```

```python
# array 열과 열의 평균
array.mean(axis=1)
```
