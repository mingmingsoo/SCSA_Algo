### K-Means

```python
# Dummy Data를 생성 (make_blobs) 

from sklearn.datasets import make_blobs # Clustering을 위한 데이터 생성

# n_samples > 데이터의 개수 (행)
# n_features > 독립변수(특징)의 개수 (열)
# centers > 클러스터의 개수 (라벨, Y)
# cluster_std > 클러스터의 표준 편차  > 클러스터간의 거리

X, Y = make_blobs(n_samples=100, n_features=2, centers=4, cluster_std=0.5, random_state=0)
X.shape, Y.shape
```

```python
Y # 정답값 (실제 클러스터링을 통한 분류가 정확하게 됬는지 확인하기 위해서 사용)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=Y) # 특징값을 가지고 시각화 (색상을 클러스터로 지정)
plt.show()
```

```python
from sklearn.cluster import KMeans # Kmeans Cluster Model, 비지도학습 > Y(x)

# n_clusters > 생성될 군집의 개수
k = KMeans(n_clusters=4).fit(X) # 군집분석은 비지도학습이므로 Y값을 넣지 않음
k
```

```python
# 2가지 방법으로 데이터가 속해있는 군집(클러스터) 정보를 받아옴

# 1. 속성값을 이용하는 경우
k.labels_ # labels > 데이터(인덱스 기준)가 속해있는 클러스터 번호를 받아옴
# fitting된 데이터의 라벨을 받아옴

# 2. 함수를 이용하는 경우 > predict() > 예측함수
X_pred = k.predict(X)
# new data의 라벨을 받아옴

k.labels_, X_pred # 결과는 동일
```

```python
# 군집의 중심
centers = k.cluster_centers_
centers
```

```python
plt.scatter(X[:, 0], X[:, 1], c=X_pred) # 색상을 예측값으로 변경
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100)
plt.show()
```

```python
# K-means 클러스터링의 문제점
from sklearn.datasets import make_moons # 클러스터링을 위한 dummy 데이터 생성 함수
# make_blobs > 선형
# make_moons > 비선형
X, Y = make_moons(200, noise=0.05, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
# Kmeans 모델에 X데이터를 fitting 시켜주세요.
k = KMeans(n_clusters=2).fit(X)
X_pred = k.predict(X)
```

```python
centers = k.cluster_centers_
centers
```

```python
# 예측결과를 시각화해주세요.
plt.scatter(X[:,0], X[:,1], c=X_pred)
plt.scatter(centers[:,0], centers[:,1], c='red', s=100)
plt.show()
```
