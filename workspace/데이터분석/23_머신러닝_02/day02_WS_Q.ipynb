### 선형 회귀 모델을 이용한 모델링

```python
# 필수 모델 및 전처리기
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
```

### Linear Regression

### 스케일링 및 다항 특성 변환을 적용한 다항 회귀 모델링
 - X 데이터에 Stanadard Scaling을 적용하세요.
 - 변환된 X에 2, 3차 다항특성 변환을 적용한 후 다시 분할하고, Lasso로 모델링하세요.
 - 적절한 alpha를 설정해보세요.
 - 학습 데이터와 테스트 데이터의 score를 각각 출력하세요.

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#다항회귀 2차 적용해서 각 알고리즘에 입력하여 성능 평가


def convert_2(X_scaled,Y, deg =2): # 다항회귀 2차 함수 만들기
    polyF = PolynomialFeatures(degree = deg)
    X_poly = polyF.fit_transform(X_scaled)
    return X_poly, Y
```

```python
X_poly,Y = convert_2(X_scaled, Y)
x_train_poly2, x_test_poly2, y_train_poly2, y_test_poly2 = train_test_split(X_poly, Y, random_state = 0, test_size = 0.25)
print(x_train_poly2.shape, x_test_poly2.shape, y_train_poly2.shape, y_test_poly2.shape)
```

```python
# Code here

# standard scaler + poly 2차 + ridge 적용
ridge = Ridge(alpha = 5)
ridge.fit(x_train_poly2,y_train_poly2)
r_train_score = ridge.score(x_train_poly2,y_train_poly2)
r_test_score = ridge.score(x_test_poly2,y_test_poly2)


# standard scaler + poly 2차 + lasso 적용
lasso = Lasso(alpha = 0.05)
lasso.fit(x_train_poly2, y_train_poly2)
l_train_score = lasso.score(x_train_poly2,y_train_poly2)
l_test_score = lasso.score(x_test_poly2,y_test_poly2)

# standard scaler + poly 2차 
model_poly = LinearRegression()
model_poly.fit(x_train_poly2, y_train_poly2)

poly_train_score = model_poly.score(x_train_poly2, y_train_poly2)
poly_test_score = model_poly.score(x_test_poly2, y_test_poly2)
```

```python
print("Ridge train score:",r_train_score)
print("Ridge test  score:",r_test_score)
print("Lasso train score:",l_train_score)
print("Lasso test  score:",l_test_score)
print("Poly train score:", poly_train_score)
print("Poly test score:", poly_test_score)
```

```python
#다항회귀 3차에 Lasso에 alpha값 : 0.01~0.3까지 간격을 0.01 주면서 성능 평가
#결과 아래와 같이 시각화 


# 다항회귀 3차 함수 만들기
def convertDataSet_3(X_scaled,Y, deg =3):
    polyF = PolynomialFeatures(degree = deg)
    X_poly = polyF.fit_transform(X_scaled)
    return X_poly, Y
```

```python
# train, test 나누기
X_poly,Y = convertDataSet_3(X_scaled, Y)
x_train_poly3, x_test_poly3, y_train_poly3, y_test_poly3 = train_test_split(X_poly, Y, random_state = 0, test_size = 0.25)
print(x_train_poly3.shape, x_test_poly3.shape, y_train_poly3.shape, y_test_poly3.shape)
```

```python
# alpha 숫자에 따라 각각의 점수를 시각화를 위해 list에 넣어줌

train_list = []
test_list = []
alpha_list = list(np.arange(0.01,0.31,0.01))
for a in np.arange(0.01,0.31,0.01): # alpha 범위
    
    lasso = Lasso(alpha = a)
    lasso.fit(x_train_poly3, y_train_poly3)
    
    l_train_score = lasso.score(x_train_poly3,y_train_poly3)
    l_test_score = lasso.score(x_test_poly3,y_test_poly3)
    
    train_list.append(l_train_score)
    test_list.append(l_test_score)
    
    print("alpha:", round(a,2 ), "train_score:", round(l_train_score,4), "test_score:", round(l_test_score,4))
```

```python
#위의 학습한 결과 시각화 

plt. figure(figsize = (8,6))
plt.plot(alpha_list, train_list, label = "train", color = "b", marker = "^")
plt.plot(alpha_list, test_list, label = "test", color = "r" , marker = ".")
plt.legend()
plt.grid(True)
plt.show()
```
