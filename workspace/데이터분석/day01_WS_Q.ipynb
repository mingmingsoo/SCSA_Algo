### 데이터 분할
 - dateset을 trainset과 testset으로 분할하세요.
 - 분할 비율은 train 8 : test 2입니다.
 - random_state = 0, 층화추출 적용

```python
# Cdoe here
from sklearn.model_selection import train_test_split

# X와 Y를 동일한 행을 추출
# 이때 test_size나 train_size를 지정하여 특정 비율로 추출이 가능하다
# stratify를 이용하여 특정 배열의 비율대로 추출도 가능하다. (ex. stratify=Y : Y의 대이터비율대로 추출)
# radom_state 를 지정하면 항상 같은 랜덤 추출 결과가 뽑힌다.
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

### KNN 모델을 이용한 모델링
 - n_neighbros는 5로 설정하세요.
 - 결과는 정확도로 trainset과 testset 모두 출력하세요.
     - 결괏값은 일치하지 않아도 됩니다.
 

```python
# Code here
from sklearn.neighbors import KNeighborsClassifier

# score은 모델마다 다른 로직으로 결과를 내는데,
# KNeighborsClassifier의 경우 (정확히 예측한 개수 / 전체 input개수) 로 정해진다.
model = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
model.score(X_train, y_train), model.score(X_test, y_test)
```

 - 데이터 재분할
 - 비율과 기타 설정은 이전과 동일합니다.

```python
# Code here
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

 - 재학습 및 결과 출력
 - 설정은 이전과 동일합니다.

```python
# Code here
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)

model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
# K의 값을 1~ 10까지 변경하면서 값을 리스트에 저장하여 출력해 보고 
# 선 그래프로 train 데이터와 테스트 데이터를 
# 출력해 보세요 
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []

for k in range(1, 11):
    model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    train_score.append(model.score(X_train, y_train))
    test_score.append(model.score(X_test, y_test))

for train, test in zip(train_score, test_score):
    print(train, test)


x = range(1, 11)
plt.plot(x, train_score, label='train')
plt.plot(x, test_score, label='test')
plt.legend()
plt.show()
```
