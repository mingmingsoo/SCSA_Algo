```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. 데이터 불러오기

```python
# 상반기 주유소 판매 데이터 불러오기

first = pd.read_csv('../data/2019년 상반기 주유소 판매가격.csv', encoding="cp949")
first
```

```python
second =  pd.read_csv('../data/2019년 하반기 주유소 판매가격.csv', encoding="cp949")
second
```

```python
# 상반기 판매가격 데이터 프레임 결측치 조회
first.isnull().sum()
```

```python
# 하반기 판매가격 데이터 프레임 결측치 조회
second.isnull().sum()
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합

df = pd.concat([first, second])
df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회

df.info()
```

```python
# 상표 컬럼의 고유값 조회

df['상표'].unique()
```

```python
# 지역 컬럼의 고유값 조회

df['지역'].unique()
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df['시'] = df['지역'].str.split(' ').str[0]
df['구'] = df['지역'].str.split(' ').str[1]
df.head()
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
df['시'] = df['시'].str.replace('서울','서울특별시')
df.head()
```

```python
# 기간 컬럼을 datetime 형식으로 변경
df['기간'] = pd.to_datetime(df['기간'].astype('str'), format="%Y%m%d")
df.head()
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성
df['년'] = df['기간'].dt.year
df['월'] = df['기간'].dt.month
df['일'] = df['기간'].dt.day
df['요일'] = df['기간'].dt.day_name()
df.head()
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values(by='휘발유', ascending=False).head(5)
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df[df['휘발유']!=0].sort_values(by='휘발유').head(5)
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
df.groupby(by='상표')['휘발유'].agg(['mean'])
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
df.groupby(by='상표')['휘발유'].agg(['mean']).plot(kind='bar')
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
df.groupby(by=['상표','셀프여부'])['휘발유'].agg(['mean'])
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 시각화하세요.
df.groupby(by=['상표','셀프여부'])['휘발유'].agg(['mean']).plot(kind='bar')
```

```python
# 셀프여부에 따라 휘발유 가격을 boxplot으로 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,8))
sns.boxplot(x="셀프여부", y="휘발유", data=df[df['휘발유']!=0], palette="Set1")
plt.show()
```

```python
# 휘발유 가격과 일반 주유소의 가격을 boxplot을 이용하여 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(12,8))
sns.boxplot(x="상표", y="휘발유", hue="셀프여부", data=df[df['휘발유']!=0], palette="Set1")
plt.show()
```

```python
# 지역별 평균 휘발유 가격을 계산하세요
df.groupby(by='구')['휘발유'].agg(['mean']).sort_values('mean', ascending=False)
```

```python
# 지역별 평균 휘발유 가격을 시각화하세요.
df.groupby(by='구')['휘발유'].agg(['mean']).sort_values('mean', ascending=False).plot(kind='bar')
```

```python
# 월별 평균 휘발유 가격을 계산하세요
df.groupby(by='월')['휘발유'].agg(['mean'])
```

```python
# 월별 평균 휘발유 가격을 시각화 하세요.
df.groupby(by='월')['휘발유'].agg(['mean']).plot(kind='bar')
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 계산하세요
df.pivot_table(index='월',columns='구', values='휘발유', aggfunc='mean')
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 시각화 하세요 (heatmap 이용)
plt.figure(figsize=(10,10))
sns.heatmap(df.pivot_table(index='월',columns='구', values='휘발유', aggfunc='mean'))
```
