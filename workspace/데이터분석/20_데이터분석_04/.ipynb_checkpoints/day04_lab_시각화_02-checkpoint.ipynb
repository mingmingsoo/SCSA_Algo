### 선 그래프(Line plot)
- plt.plot() 함수 사용
- plt.plot([x], y, [fmt], [data]) 함수를 사용하여 x, y 축을 갖는 그래프 생성
    - x, y : array-like or scalar
    - fmt : str, optional, A format string, e.g. 'ro--' for red circles with dashed line
    - 'ro--' : fmtcolor='red', marker='o', linestyle='dashed'
    - data : indexable object, optional, DataFrame 같은 객체를 주면, x, y를 column 이름으로 줄 수 있음
- plot 설명에서 색, marker 등에 대한 정보 확인
- https://matplotlib.org/2.0.2/examples/color/named_colors.html
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

<br/>
<br/>

#### 하나의 선 그리기
- plt.plot(y) : x는 [0, 1, 2, ..] 자동생성
- plt.plot(x, y) : x, y 둘 다 지정, 개수가 동일해야 함
- plt.plt(x, y, fmt) : 선의 스타일 지정

````python
# (0,0), (1,2), (2,1) 세 점을 연결한 선 그래프

x = [0,1,2]
y = [0,2,1]

plt.plot(x, y)

plt.show()
````

````python
# (0,0), (1,2), (2,1) 세 점을 연결한 선 그래프
# (0,2), (1,0), (2,2) 세 점을 연결한 선 그래프

x = [0,1,2]
y1 = [0,2,1]
y2 = [2,0,2]

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
````

````python
# (0,0), (1,2), (2,1) 세 점을 연결한 선 그래프 (red)
# (0,2), (1,0), (2,2) 세 점을 연결한 선 그래프 (black)

x = [0,1,2]
y1 = [0,2,1]
y2 = [2,0,2]

plt.plot(x, y1, color = 'red')
plt.plot(x, y2, color = 'black')


plt.show()
````

````python
# X의 범위가 -6부터 6까지의 sin 그래와 cos 그래프를 그리세요.

x = np.linspace(-6,6,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, c = 'r')
plt.plot(x, y2, c = 'k')

plt.show()
````

#### 범례 표시
- Legend(범례), 범례 설정을 위해 그래프에 label이 있어야 함
- 방법1
        plt.plot(x, y, style, label='text')
        plt.legend(loc='text')

- 방법2
        plt.plot에 label 생략 
        plt.legend(labels=[label1, label2 ...],loc='text')사용
        (순서, 개수 동일)

- loc : upper left, upper center, upper right, center left, center, center right, lower left, lower right, best, right 등


````python
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm

# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows 사용자의 경우
# plt.rcParams['font.family'] = 'AppleGothic'  # macOS 사용자의 경우
# plt.rcParams['font.family'] = 'NanumGothic'  # Linux 사용자의 경우

# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
````

````python
# 위의 그래프에 범례(legend)를 추가하세요.




# X의 범위가 -6부터 6까지의 sin 그래와 cos 그래프를 그리세요.

x = np.linspace(-6,6,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, c = 'r',label ="sin")
plt.plot(x, y2, c = 'k',label ="cos")

plt.legend(loc = "upper right")

plt.show()


````

#### x,y 축 레이블 표시
- plt.xlabel(str, fontsize) : x 축에 대한 레이블 표시
- plt.ylabel(str, fontsize) : y 축에 대한 레이블 표시
- plt.title(str, fontsize) : plot에 대한 제목

````python
# 위의 그래프에 x축을 'X', y축을 'Y'로 표시하세요.

x = np.linspace(-6,6,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, c = 'r',label ="sin")
plt.plot(x, y2, c = 'k',label ="cos")

plt.legend(loc = "upper right")

plt.xlabel("X",fontsize = 8)
plt.ylabel("Y",fontsize = 8)
plt.title("Sine and Cosine",fontsize = 15)
plt.show()








````

#### x, y 축 특정 영역만 표시
- Limitation : 전체 데이터 중에 특정 영역만 표시
    - plt.xlim(min, max)
    - plt.ylim(min, max)

````python
# 위의 그래프에 x축의 범위를 -4 ~ 4로 한정하세요.



x = np.linspace(-6,6,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, c = 'r',label ="sin")
plt.plot(x, y2, c = 'k',label ="cos")

plt.legend(loc = "upper right")

plt.xlim(-4,4)

plt.xlabel("X",fontsize = 8)
plt.ylabel("Y",fontsize = 8)
plt.title("Sine and Cosine",fontsize = 15)
plt.show()

````

### 5. Figure Size(단위:inch)
- Matplotlib에서 그래프(Axes, plot)는 Figure 객체내에 존재함
- 그래프를 위한 새로운 Figure는 plt.figure() 함수를 통해 생성함
- plt.figure(번호, figsize=크기, dpi=1인치당 픽셀)
- 크기 단위 : 인치, 기본 1인치 = 72 픽셀 (dpi를 사용해 조절 가능)

````python
# 위의 그래프의 크기를 가로 12인치, 세로 6인치로 변경하세요.


x = np.linspace(-6,6,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize = (12,6))

plt.plot(x, y1, c = 'r',label ="sin")
plt.plot(x, y2, c = 'k',label ="cos")

plt.legend(loc = "upper right")

plt.xlim(-4,4)

plt.xlabel("X",fontsize = 8)
plt.ylabel("Y",fontsize = 8)
plt.title("Sine and Cosine",fontsize = 15)
plt.show()

````

### 3. Scatter (산점도, 산포도, 분산 형 차트)
- 산점도, 산포도, 분산 형 차트
- 두 데이터간의 관계를 보기 위해 사용함
- plt.scatter(x, y, s, c, alpha)
   - s : scale
   - c : color
   - alpha : 투명도(0~1)
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html   
- line style은 링크 참고
  - https://matplotlib.org/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py

````python
# x, y를 산점도 그래프로 그리세요.
x = np.arange(50)
y = x + np.random.randn(50)*6


plt.scatter(x,y,s=300,c="r",alpha = 0.5,marker = "$😍$")

plt.show()
````

````python
# 위의 그래프에 z 값을 기준으로 색깔을 지정하세요.
x = np.arange(50)
y = x + np.random.randn(50)*6
z = [0] * 25 + [1] * 25

plt.scatter(x,y,s=300,c=z,alpha = 0.5,marker = "$😍$")


plt.show()
````

### bar 그래프 (범주형)
- 여러 대상의 양적의 비교가 필요할 때 사용
- 시간의 흐름에 따른 비교도 할 수 있지만, 변화의 흐름 보다는 변화의 크기에 집중이 필요할 때 사용함
- plt.bar(x, y) : 수직바
- plt.barh(x, y) : 수평바
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.barh.html

````python
# name 별 values 값을 막대그래프로 그리세요.
names = ['A', 'B', 'C']
values = [50, 30, 100]



plt.bar(names,values)
plt.show()
````

````python
# name 별 values 값을 수평막대그래프로 그리세요.
names = ['A', 'B', 'C']
values = [50, 30, 100]

plt.barh(names,values)

plt.show()
````

### 히스토그램 (연속형)
- 변수가 하나인 단변수 데이터의 빈도수를 그래프로 표현(일변량 그래프)
- 히스토그램을 그리기 위한 hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받음
- 데이터 집계 결과를 반환
- n, bins, patches = plt.hist(x, bins, range)
   - x : 데이터
   - bins : 계급
   - range: 대상 범위, (min, max)
   - n : 도수 
   - patches : 계급별 히스토그램 좌표와 높이
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html  

````python
# x의 구간별 데이터의 분포를 확인하세요. (구간을 50개로 나누세요.)
x = np.random.randn(1000)

plt.hist(x,bins = 50)

plt.show()
````

## Tips Data를 활용한 시각화 실습

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = sns.load_dataset("tips")
df.head(5)
````

#### boxplot 

````python
# tips의 'total_bill' 의 분포를 boxplot으로 확인함 (x로 지정)
plt.figure(figsize=(8, 6))

sns.boxplot(x = df["total_bill"])

plt.show()
````

````python
# 특정 카테고리에 대한 값의 분포는 X,Y를 지정하여 확인 가능
# x, y, data 사용
plt.figure(figsize=(8, 6))

sns.boxplot(y = "total_bill", x = "day", data = df)


plt.show()
````

````python
# hue 옵션을 통해 카테고리 별 데이터 시각화 가능
# palette 옵션을 통해 색상 지정 가능 ('Set3' 사용)
plt.figure(figsize=(8, 6))


sns.boxplot(y = "total_bill", x = "day", hue = "smoker", data = df, palette ="Set3")


plt.show()
````

#### heatmap
- 두 개의 카테고리 값에 의한 실수의 변화를 보고자 할 경우에 사용
- 표출하고자 하는 데이터를 먼저 데이터 테이블 형태로 만들어야 함

````python
flights = sns.load_dataset("flights")
flights.head(5)
````

````python
flights = pd.pivot_table(flights, index = 'month', columns = 'year', values = 'passengers')
flights
````

````python
plt.figure(figsize=(10, 8))

sns.heatmap(flights)

plt.show()
````

````python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
````

#### pairplot 
- 데이터 분포 및 변수간 관계를 분석

````python
iris = sns.load_dataset("iris")
iris.head(10)
````

````python
# pairplot을 이용하여 변수(컬럼)간 관계와 분포 확인 가능
plt.figure(figsize = (10,12))
sns.pairplot(data = iris)

plt.show()
````

````python
# hue 옵션을 이용하여 카테고리별 변수의 관계와 분포 확인 가능
sns.pairplot(data = iris, hue = "species")

plt.show()
````
